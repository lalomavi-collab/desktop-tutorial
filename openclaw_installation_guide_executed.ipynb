```python
from IPython.display import display, HTML

# Color scheme
COLORS = {
    "title_bg": "#2C3E50",
    "title_text": "#FFFFFF",
    "body_bg": "#FAFAFA",
    "body_text": "#2C3E50",
    "code_bg": "#1E1E1E",
    "code_text": "#D4D4D4",
    "accent": "#27AE60",
    "subtitle": "#7F8C8D",
    "border": "#BDC3C7",
    "slide_num": "#95A5A6",
}

# Icons per slide
ICONS = [
    "&#128640;",  # 0 - rocket (title)
    "&#129302;",  # 1 - robot (what is)
    "&#128221;",  # 2 - checklist (prereqs)
    "&#9889;",    # 3 - lightning (one-liner)
    "&#128230;",  # 4 - package (npm)
    "&#128051;",  # 5 - whale (docker)
    "&#129433;",  # 6 - llama (ollama)
    "&#9729;",    # 7 - cloud
    "&#128295;",  # 8 - wrench (setup)
    "&#128260;",  # 9 - arrows (update)
    "&#128274;",  # 10 - lock (security)
    "&#127911;",  # 11 - headphones (notebooklm)
    "&#127912;",  # 12 - artist (customize)
    "&#127937;",  # 13 - checkered flag (summary)
]

def format_content(text):
    """Convert plain text content to styled HTML"""
    lines = text.split("\n")
    html_lines = []
    in_code = False
    
    for line in lines:
        stripped = line.strip()
        if not stripped:
            html_lines.append("<div style='height:12px'></div>")
            continue
        
        # Detect command lines
        is_command = any(stripped.startswith(cmd) for cmd in [
            "curl ", "npm ", "pnpm ", "docker ", "ollama ", "openclaw ",
            "node ", "# ", "docs.", "notebooklm."
        ])
        
        if is_command:
            html_lines.append(
                f"<div style='background:{COLORS['code_bg']};color:{COLORS['code_text']};"
                f"padding:8px 16px;border-radius:6px;font-family:monospace;"
                f"font-size:14px;margin:4px 0;direction:ltr;text-align:left'>"
                f"<code>{stripped}</code></div>"
            )
        elif stripped.startswith("•"):
            bullet_text = stripped[1:].strip()
            if bullet_text.startswith("—") or bullet_text.startswith("  "):
                # Sub-bullet
                html_lines.append(
                    f"<div style='padding:3px 0 3px 40px;font-size:15px;color:{COLORS['subtitle']}'>"
                    f"&#8594; {bullet_text.lstrip('— ')}</div>"
                )
            else:
                html_lines.append(
                    f"<div style='padding:4px 0 4px 20px;font-size:16px'>"
                    f"<span style='color:{COLORS['accent']};font-weight:bold'>&#9679;</span> "
                    f"{bullet_text}</div>"
                )
        elif stripped[0].isdigit() and "." in stripped[:3]:
            # Numbered list
            html_lines.append(
                f"<div style='padding:4px 0 4px 20px;font-size:16px'>"
                f"<span style='color:{COLORS['accent']};font-weight:bold'>{stripped[:2]}</span>"
                f"{stripped[2:]}</div>"
            )
        elif stripped.endswith(":"):
            # Section header
            html_lines.append(
                f"<div style='padding:8px 0 4px 0;font-size:15px;font-weight:bold;"
                f"color:{COLORS['accent']}'>{stripped}</div>"
            )
        elif "—" in stripped and stripped.startswith("  "):
            html_lines.append(
                f"<div style='padding:3px 0 3px 40px;font-size:15px;color:{COLORS['subtitle']}'>"
                f"&#8594; {stripped.strip().lstrip('— ')}</div>"
            )
        else:
            html_lines.append(f"<div style='padding:3px 0;font-size:16px'>{stripped}</div>")
    
    return "\n".join(html_lines)

# Build full slideshow HTML
all_slides_html = ""
for i, slide_data in enumerate(slides_content):
    icon = ICONS[i] if i < len(ICONS) else "&#128196;"
    content_html = format_content(slide_data["content"])
    
    if i == 0:
        # Title slide - centered, bigger
        slide_html = f"""
        <div style="
            background: linear-gradient(135deg, {COLORS['title_bg']} 0%, #34495E 50%, {COLORS['accent']} 100%);
            border-radius: 16px;
            padding: 60px 40px;
            margin: 20px 0;
            text-align: center;
            box-shadow: 0 8px 32px rgba(0,0,0,0.15);
            min-height: 350px;
            display: flex;
            flex-direction: column;
            justify-content: center;
        ">
            <div style="font-size:64px;margin-bottom:20px">{icon}</div>
            <h1 style="color:white;font-size:42px;margin:0 0 20px 0;font-weight:700;
                text-shadow:2px 2px 4px rgba(0,0,0,0.3)">{slide_data['title']}</h1>
            <div style="color:rgba(255,255,255,0.85);font-size:20px;line-height:1.6">
                {slide_data['content'].replace(chr(10), '<br>')}
            </div>
            <div style="color:rgba(255,255,255,0.5);font-size:14px;margin-top:30px">
                Slide {i+1} / {len(slides_content)}
            </div>
        </div>
        """
    else:
        slide_html = f"""
        <div style="
            background: white;
            border-radius: 16px;
            margin: 20px 0;
            box-shadow: 0 4px 20px rgba(0,0,0,0.08);
            border: 1px solid {COLORS['border']};
            overflow: hidden;
            min-height: 350px;
        ">
            <!-- Slide header -->
            <div style="
                background: linear-gradient(135deg, {COLORS['title_bg']}, #34495E);
                padding: 20px 30px;
                display: flex;
                align-items: center;
                justify-content: space-between;
            ">
                <div style="display:flex;align-items:center;gap:15px">
                    <span style="font-size:36px">{icon}</span>
                    <h2 style="color:white;margin:0;font-size:26px;font-weight:600">
                        {slide_data['title']}
                    </h2>
                </div>
                <span style="color:rgba(255,255,255,0.5);font-size:14px">
                    {i+1} / {len(slides_content)}
                </span>
            </div>
            <!-- Slide body -->
            <div style="padding:30px;color:{COLORS['body_text']};direction:rtl;text-align:right">
                {content_html}
            </div>
        </div>
        """
    all_slides_html += slide_html

# Wrap everything
full_html = f"""
<div style="max-width:900px;margin:0 auto;font-family:'Segoe UI',Tahoma,Geneva,Verdana,sans-serif">
    <div style="text-align:center;padding:10px 0;color:{COLORS['slide_num']};font-size:13px">
        OpenClaw Installation Guide &mdash; {len(slides_content)} Slides
    </div>
    {all_slides_html}
</div>
"""

display(HTML(full_html))
```

## יצירת מצגת PowerPoint

```python
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor
from pptx.enum.text import PP_ALIGN

prs = Presentation()
prs.slide_width = Inches(13.333)
prs.slide_height = Inches(7.5)

# Pastel color scheme
TITLE_COLOR = RGBColor(0x2C, 0x3E, 0x50)   # Dark blue-gray
ACCENT_COLOR = RGBColor(0x27, 0xAE, 0x60)  # Green
TEXT_COLOR = RGBColor(0x34, 0x49, 0x5E)     # Medium blue-gray
CODE_COLOR = RGBColor(0xE7, 0x4C, 0x3C)    # Red for commands

for i, slide_data in enumerate(slides_content):
    if i == 0:
        slide = prs.slides.add_slide(prs.slide_layouts[0])
    else:
        slide = prs.slides.add_slide(prs.slide_layouts[1])

    # Title
    title_shape = slide.shapes.title
    title_shape.text = slide_data["title"]
    for para in title_shape.text_frame.paragraphs:
        para.font.size = Pt(36 if i == 0 else 30)
        para.font.bold = True
        para.font.color.rgb = TITLE_COLOR

    # Content
    if len(slide.placeholders) > 1:
        body = slide.placeholders[1]
        body.text = slide_data["content"]
        for para in body.text_frame.paragraphs:
            para.font.size = Pt(18 if i == 0 else 16)
            para.font.color.rgb = TEXT_COLOR
            para.space_after = Pt(8)

pptx_path = "openclaw_installation_guide.pptx"
prs.save(pptx_path)
import os
print(f"✅ Presentation saved: {pptx_path} ({os.path.getsize(pptx_path)/1024:.1f} KB)")
print(f"   Total slides: {len(prs.slides)}")
```

## יצירת PDF של השקפים

```python
from reportlab.lib.pagesizes import landscape, A4
from reportlab.lib.colors import HexColor
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, PageBreak
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.units import inch
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont
import os

pdf_path = "openclaw_installation_guide.pdf"

doc = SimpleDocTemplate(
    pdf_path,
    pagesize=landscape(A4),
    topMargin=0.8*inch,
    bottomMargin=0.5*inch,
    leftMargin=1*inch,
    rightMargin=1*inch
)

styles = getSampleStyleSheet()
title_style = ParagraphStyle(
    'SlideTitle', parent=styles['Heading1'],
    fontSize=26, spaceAfter=20,
    textColor=HexColor('#2C3E50'),
    alignment=1
)
body_style = ParagraphStyle(
    'SlideBody', parent=styles['Normal'],
    fontSize=14, leading=20,
    textColor=HexColor('#34495E'),
    spaceAfter=8
)

story = []
for i, slide_data in enumerate(slides_content):
    if i > 0:
        story.append(PageBreak())
    story.append(Spacer(1, 0.3*inch))
    story.append(Paragraph(slide_data["title"], title_style))
    story.append(Spacer(1, 0.2*inch))
    for line in slide_data["content"].split("\n"):
        if line.strip():
            story.append(Paragraph(line.strip(), body_style))

doc.build(story)
print(f"✅ PDF saved: {pdf_path} ({os.path.getsize(pdf_path)/1024:.1f} KB)")
print(f"   Pages: {len(slides_content)}")
```

## יצירת פודקאסט (אודיו)

```python
import subprocess, os

# Podcast script about OpenClaw installation
podcast_script = """
Welcome to this episode about OpenClaw, the open-source AI agent framework.

OpenClaw is a self-hosted personal AI assistant created by Peter Steinberger.
It has over 196,000 GitHub stars and 600 contributors, making it one of the fastest growing open-source projects ever.

Let's talk about installation. There are five main ways to install OpenClaw.

First, the one-liner method. This is the recommended approach. It automatically installs Node.js and all dependencies with a single command.

Second, npm. Just run npm install dash g openclaw. You'll need Node.js version 22.12 or higher.

Third, Docker. This gives you full process isolation and is best for server deployments. Pull the latest image and run it.

Fourth, Ollama integration. Since Ollama 0.17, you can run OpenClaw on your own hardware with a single command.

Fifth, cloud deployment. DigitalOcean and Hostinger offer one-click deployments.

After installation, run the onboarding wizard. It configures your gateway, communication channels, skills, and workspace.

For updates, use npm update dash g openclaw, or docker pull the latest image.
Always run openclaw security audit dash dash deep after updating.

Security is critical. The Canvas Host defaults to binding all interfaces, so anyone on your network can reach it.
Deploy on a VPS, not your personal computer.

Finally, you can use Google NotebookLM to turn this guide into an engaging podcast.
Upload the PDF, click Audio Overview, and NotebookLM creates a conversation between two AI hosts.
You can customize the tone, focus, and structure.

That's your complete guide to OpenClaw installation and updates. Thanks for listening!
"""

# Generate WAV with espeak-ng
wav_path = "openclaw_podcast.wav"
result = subprocess.run(
    ['espeak-ng', '-w', wav_path, '-s', '145', '-p', '50', podcast_script.strip()],
    capture_output=True, text=True
)

if result.returncode == 0:
    wav_size = os.path.getsize(wav_path)
    print(f"✅ WAV audio saved: {wav_path} ({wav_size/1024:.1f} KB)")
    
    # Compress to MP3
    mp3_path = "openclaw_podcast.mp3"
    result2 = subprocess.run(
        ['ffmpeg', '-y', '-i', wav_path,
         '-codec:a', 'libmp3lame', '-b:a', '64k', '-ac', '1',
         mp3_path],
        capture_output=True, text=True
    )
    if result2.returncode == 0:
        mp3_size = os.path.getsize(mp3_path)
        reduction = (1 - mp3_size / wav_size) * 100
        print(f"✅ MP3 podcast saved: {mp3_path} ({mp3_size/1024:.1f} KB)")
        print(f"   Compression: {reduction:.0f}% reduction")
    else:
        print(f"❌ ffmpeg error: {result2.stderr[-200:]}")
else:
    print(f"❌ espeak-ng error: {result.stderr}")
```
