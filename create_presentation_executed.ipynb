## שלב 1: בחירת נושא וסרטון

לדוגמה, נבחר סרטון על **"מבוא לבינה מלאכותית"**:

- נושא: מבוא לבינה מלאכותית (AI)
- מקור: סרטון לימודי מ-YouTube
- מטרה: ליצור מצגת מסכמת של 5 שקפים

```python
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.enum.text import PP_ALIGN
from pptx.dml.color import RGBColor

print("python-pptx loaded successfully!")
```

## שלב 2: הגדרת תוכן המצגת

נגדיר את התוכן של כל שקף — כאילו סיכמנו סרטון לימודי על מבוא ל-AI:

```python
slides_content = [
    {
        "title": "מבוא לבינה מלאכותית",
        "content": "סיכום סרטון לימודי\nהוכן באופן אוטומטי באמצעות Python"
    },
    {
        "title": "מהי בינה מלאכותית?",
        "content": (
            "• תחום במדעי המחשב העוסק ביצירת מערכות חכמות\n"
            "• מטרה: לחקות יכולות חשיבה אנושיות\n"
            "• כולל: למידת מכונה, עיבוד שפה טבעית, ראייה ממוחשבת"
        )
    },
    {
        "title": "סוגי בינה מלאכותית",
        "content": (
            "• AI צר (Narrow AI) — מתמחה במשימה אחת\n"
            "• AI כללי (General AI) — יכולת חשיבה רחבה כמו אדם\n"
            "• AI על-אנושי (Super AI) — עולה על יכולת אנושית (תיאורטי)"
        )
    },
    {
        "title": "יישומים בחיי היום-יום",
        "content": (
            "• עוזרים וירטואליים (Siri, Alexa, Claude)\n"
            "• המלצות תוכן (Netflix, Spotify)\n"
            "• רכבים אוטונומיים\n"
            "• אבחון רפואי"
        )
    },
    {
        "title": "סיכום ומבט לעתיד",
        "content": (
            "• AI משנה את העולם בקצב מואץ\n"
            "• חשוב להבין את ההזדמנויות והאתגרים\n"
            "• הלמידה המתמשכת היא המפתח להצלחה\n"
            "\nתודה על הצפייה!"
        )
    }
]

print(f"Defined {len(slides_content)} slides")
```

## שלב 3: יצירת המצגת

```python
prs = Presentation()
prs.slide_width = Inches(13.333)
prs.slide_height = Inches(7.5)

for i, slide_data in enumerate(slides_content):
    if i == 0:
        layout = prs.slide_layouts[0]  # Title slide
    else:
        layout = prs.slide_layouts[1]  # Title + content

    slide = prs.slides.add_slide(layout)

    # Set title
    slide.shapes.title.text = slide_data["title"]
    for paragraph in slide.shapes.title.text_frame.paragraphs:
        paragraph.font.size = Pt(36)
        paragraph.font.bold = True

    # Set content
    if len(slide.placeholders) > 1:
        body = slide.placeholders[1]
        body.text = slide_data["content"]
        for paragraph in body.text_frame.paragraphs:
            paragraph.font.size = Pt(20)

output_path = "presentation_ai_intro.pptx"
prs.save(output_path)
print(f"Presentation saved: {output_path}")
print(f"Total slides: {len(prs.slides)}")
```
